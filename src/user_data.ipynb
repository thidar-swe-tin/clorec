```python
import pandas as pd
```

```python
user_reviews_df = pd.read_csv("../data/orders.csv")
user_reviews_df.head()
```

```python
user_review_rating_df = user_reviews_df[["user_id","product_id"]]
user_review_rating_df.head()
```

```python
user_rating_df = user_review_rating_df.groupby("user_id")["product_id"].count().reset_index()
user_rating_df = user_rating_df.sort_values(by='product_id').reset_index(drop=True)
user_rating_df = user_rating_df[user_rating_df.product_id==10].reset_index(drop=True)
user_rating_df = user_rating_df[:10]
user_rating_df = user_rating_df.drop(['product_id'], axis=1)
user_rating_df.to_csv('../data/users.csv', index=False)
user_rating_df
```
